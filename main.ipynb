```python
import json

def extract_unique_types_from_json(data):
    """
    –ò–∑–≤–ª–µ–∫–∞–µ—Ç —É–Ω–∏–∫–∞–ª—å–Ω—ã–µ —Ç–∏–ø—ã –∏–∑ JSON-–¥–∞–Ω–Ω—ã—Ö.

    :param data: JSON-–¥–∞–Ω–Ω—ã–µ –≤ –≤–∏–¥–µ —Å–ª–æ–≤–∞—Ä—è.
    :return: –ú–Ω–æ–∂–µ—Å—Ç–≤–æ —É–Ω–∏–∫–∞–ª—å–Ω—ã—Ö —Ç–∏–ø–æ–≤.
    """
    unique_types = set()
    for entity in data.get("entities", []):
        types = entity.get("types", [])
        for t in types:
            unique_types.add(t)  # –î–æ–±–∞–≤–ª—è–µ–º —Ç–∏–ø—ã –±–µ–∑ –∏–∑–º–µ–Ω–µ–Ω–∏—è —Ä–µ–≥–∏—Å—Ç—Ä–∞
    return unique_types

# –ü—Ä–∏–º–µ—Ä –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏—è
data = {
    "text": "–í–ª–∞—Å—Ç–∏ —Å–º–æ–≥—É—Ç –∑–∞–ø—Ä–∞—à–∏–≤–∞—Ç—å —É –±–∏–∑–Ω–µ—Å–∞ –¥–∞–Ω–Ω—ã–µ –ø–æ –ª—é–±–æ–º—É –ø–æ–≤–æ–¥—É. –ú–∏–Ω—Ü–∏—Ñ—Ä—ã —Ä–∞–∑—Ä–∞–±–æ—Ç–∞–ª–æ –ø—Ä–æ–µ–∫—Ç —Ä–∞—Å–ø–æ—Ä—è–∂–µ–Ω–∏—è –ø—Ä–∞–≤–∏—Ç–µ–ª—å—Å—Ç–≤–∞, –¥–µ—Ç–∞–ª–∏–∑–∏—Ä—É—é—â–∏–π —Å–ª—É—á–∞–∏ –∑–∞–ø—Ä–æ—Å–∞ –æ–±–µ–∑–ª–∏—á–µ–Ω–Ω—ã—Ö –ø–µ—Ä—Å–æ–Ω–∞–ª—å–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö –¥–ª—è —Å–æ–∑–¥–∞–Ω–∏—è –≥–æ—Å—Å–æ–∑–µ—Ä–∞ –¥–∞–Ω–Ω—ã—Ö. –ö–æ–º–ø–∞–Ω–∏–∏ –æ–ø–∞—Å–∞—é—Ç—Å—è –∑–Ω–∞—á–∏—Ç–µ–ª—å–Ω—ã—Ö —Ä–∞—Å—Ö–æ–¥–æ–≤ –Ω–∞ –æ–±–µ–∑–ª–∏—á–∏–≤–∞–Ω–∏–µ –∏ –Ω–µ–ø—Ä–æ–∑—Ä–∞—á–Ω–æ—Å—Ç–∏ –∑–∞–ø—Ä–æ—Å–æ–≤, –∞ —Ç–∞–∫–∂–µ —Ä–∏—Å–∫–∞ –æ–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –ø—Ä–∏–Ω–∞–¥–ª–µ–∂–Ω–æ—Å—Ç–∏ –¥–∞–Ω–Ω—ã—Ö –∫–æ–Ω–∫—Ä–µ—Ç–Ω–æ–º—É —á–µ–ª–æ–≤–µ–∫—É.",
    "entities": [
        {"entity": "–í–ª–∞—Å—Ç–∏", "types": ["Organization"]},
        {"entity": "–±–∏–∑–Ω–µ—Å–∞", "types": ["Organization"]},
        {"entity": "–ú–∏–Ω—Ü–∏—Ñ—Ä—ã", "types": ["Organization"]},
        {"entity": "–ø—Ä–∞–≤–∏—Ç–µ–ª—å—Å—Ç–≤–∞", "types": ["Organization", "Government"]},
        {"entity": "–≥–æ—Å—Å–æ–∑–µ—Ä–∞ –¥–∞–Ω–Ω—ã—Ö", "types": ["Organization", "Database"]},
        {"entity": "–ö–æ–º–ø–∞–Ω–∏–∏", "types": ["Organization"]}
    ],
    "label": [
        "DATA",
        "ORGANIZATION",
        "PERSON",
        "DOCUMENT",
        "SYSTEM",
        "LAW"
    ]
}

# –ò–∑–≤–ª–µ—á–µ–Ω–∏–µ —É–Ω–∏–∫–∞–ª—å–Ω—ã—Ö —Ç–∏–ø–æ–≤
unique_types = extract_unique_types_from_json(data)

# –í—ã–≤–æ–¥ —É–Ω–∏–∫–∞–ª—å–Ω—ã—Ö —Ç–∏–ø–æ–≤
print("–£–Ω–∏–∫–∞–ª—å–Ω—ã–µ —Ç–∏–ø—ã:", (unique_types))
```

```python
from pydantic import BaseModel
from typing import List
from openai import OpenAI

# –ò–Ω–∏—Ü–∏–∞–ª–∏–∑–∞—Ü–∏—è –∫–ª–∏–µ–Ω—Ç–∞ OpenAI
client = OpenAI(base_url="http://localhost:11434/v1", api_key="ollama")

# –û–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ —Å—Ö–µ–º—ã –¥–ª—è Entity
class Entity(BaseModel):
    entity: str
    types: List[str]

# –û–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ —Å—Ö–µ–º—ã –¥–ª—è Entry
class Entry(BaseModel):
    text: str
    entities: List[Entity]

try:
    # –û—Ç–ø—Ä–∞–≤–∫–∞ –∑–∞–ø—Ä–æ—Å–∞ –∫ –º–æ–¥–µ–ª–∏
    completion = client.chat.completions.create(
        temperature=0,
        model="qwen2.5:7b",
        messages=[
            {"role": "user", "content": "Extract entities from the text: 'John is a developer who works at OpenAI. He lives in San Francisco.' Return the entities in JSON format."}
        ],
    )

    # –û–±—Ä–∞–±–æ—Ç–∫–∞ –æ—Ç–≤–µ—Ç–∞
    if completion.choices:
        response_message = completion.choices[0].message
        if response_message.content:
            # –ü–∞—Ä—Å–∏–Ω–≥ JSON-–æ—Ç–≤–µ—Ç–∞ –≤ –æ–±—ä–µ–∫—Ç Entry
            parsed_response = Entry.parse_raw(response_message.content)
            print("Parsed response:", parsed_response)
        else:
            print("No content in the response.")
    else:
        print("No choices in the response.")
except Exception as e:
    print(f"Error: {e}")
```

```python
def split_text_in_memory(filepath, chunk_size_chars=1000, overlap_chars=200):
    """–ß–∏—Ç–∞–µ—Ç –≤–µ—Å—å —Ñ–∞–π–ª –≤ –ø–∞–º—è—Ç—å –∏ —Ä–∞–∑–±–∏–≤–∞–µ—Ç –µ–≥–æ –Ω–∞ —á–∞—Å—Ç–∏ —Å –ø–µ—Ä–µ–∫—Ä—ã—Ç–∏–µ–º."""
    try:
        with open(filepath, 'r', encoding='utf-8') as f:
            text = f.read()

        start = 0
        chunks = []
        while start < len(text):
            end = min(start + chunk_size_chars, len(text))
            
            # –ù–∞—Ö–æ–¥–∏–º –ø–æ—Å–ª–µ–¥–Ω–∏–π –ø–µ—Ä–µ–Ω–æ—Å —Å—Ç—Ä–æ–∫–∏ –≤ –ø—Ä–µ–¥–µ–ª–∞—Ö overlap
            if end + overlap_chars < len(text):
              overlap = text[end:end + overlap_chars]
              newline_pos = overlap.rfind('\n')

              if newline_pos != -1:
                end = end + newline_pos + 1
            
            chunk = text[start:end]
            chunks.append(chunk)
            start = end

        return chunks

    except FileNotFoundError:
        print(f"–§–∞–π–ª {filepath} –Ω–µ –Ω–∞–π–¥–µ–Ω.")
        return [] # –í–æ–∑–≤—Ä–∞—â–∞–µ–º –ø—É—Å—Ç–æ–π —Å–ø–∏—Å–æ–∫ –≤ —Å–ª—É—á–∞–µ –æ—à–∏–±–∫–∏
    except Exception as e:
        print(f"–ü—Ä–æ–∏–∑–æ—à–ª–∞ –æ—à–∏–±–∫–∞: {e}")
        return []



# –ü—Ä–∏–º–µ—Ä –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏—è:
filepath = "input_data/4pages.txt" # –ó–∞–º–µ–Ω–∏—Ç–µ –Ω–∞ –≤–∞—à –ø—É—Ç—å
chunk_size = 1000
overlap = 200

chunks = split_text_in_memory(filepath, chunk_size, overlap)

for i, chunk in enumerate(chunks):
    print(f"–û–±—Ä–∞–±–æ—Ç–∫–∞ —á–∞—Å—Ç–∏ {i+1}:")
    # ... –≤–∞—à –∫–æ–¥ –¥–ª—è –æ–±—Ä–∞–±–æ—Ç–∫–∏ chunk ...
    print(chunk)
    print('-'*10)
    print("\n" * 4)
```
