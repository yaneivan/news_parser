```python
import json

def extract_unique_types_from_json(data):
    """
    Извлекает уникальные типы из JSON-данных.

    :param data: JSON-данные в виде словаря.
    :return: Множество уникальных типов.
    """
    unique_types = set()
    for entity in data.get("entities", []):
        types = entity.get("types", [])
        for t in types:
            unique_types.add(t)  # Добавляем типы без изменения регистра
    return unique_types

# Пример использования
data = {
    "text": "Власти смогут запрашивать у бизнеса данные по любому поводу. Минцифры разработало проект распоряжения правительства, детализирующий случаи запроса обезличенных персональных данных для создания госсозера данных. Компании опасаются значительных расходов на обезличивание и непрозрачности запросов, а также риска определения принадлежности данных конкретному человеку.",
    "entities": [
        {"entity": "Власти", "types": ["Organization"]},
        {"entity": "бизнеса", "types": ["Organization"]},
        {"entity": "Минцифры", "types": ["Organization"]},
        {"entity": "правительства", "types": ["Organization", "Government"]},
        {"entity": "госсозера данных", "types": ["Organization", "Database"]},
        {"entity": "Компании", "types": ["Organization"]}
    ],
    "label": [
        "DATA",
        "ORGANIZATION",
        "PERSON",
        "DOCUMENT",
        "SYSTEM",
        "LAW"
    ]
}

# Извлечение уникальных типов
unique_types = extract_unique_types_from_json(data)

# Вывод уникальных типов
print("Уникальные типы:", (unique_types))
```

```python
import pandas as pd
import os
from load_data import load_texts


# File path
file_path = 'texts.csv'

# Specific column names
column_names = ['full text', 'summary', 'LLM json', 'LLM summary json', 'metadata']

# Check if the file exists
if os.path.exists(file_path):
    existing_df = pd.read_csv(file_path)
else:
    existing_df = pd.DataFrame(columns=column_names)

# List of new texts to add
new_texts = (load_texts('forbes/result.json'))

# Filter out texts that are already in the DataFrame
new_texts_to_add = [text for text in new_texts if text not in existing_df['full text'].values]

# Create a DataFrame from the new texts to add
new_df = pd.DataFrame({'full text': new_texts_to_add})

# Append the new DataFrame to the existing DataFrame
updated_df = pd.concat([existing_df, new_df], ignore_index=True)

# Save the updated DataFrame back to the CSV file
updated_df.to_csv(file_path, index=False)

print(updated_df)
```

```python
from pydantic import BaseModel
from typing import List
from openai import OpenAI

# Инициализация клиента OpenAI
client = OpenAI(base_url="http://localhost:11434/v1", api_key="ollama")

# Определение схемы для Entity
class Entity(BaseModel):
    entity: str
    types: List[str]

# Определение схемы для Entry
class Entry(BaseModel):
    text: str
    entities: List[Entity]

try:
    # Отправка запроса к модели
    completion = client.chat.completions.create(
        temperature=0,
        model="qwen2.5:7b",
        messages=[
            {"role": "user", "content": "Extract entities from the text: 'John is a developer who works at OpenAI. He lives in San Francisco.' Return the entities in JSON format."}
        ],
    )

    # Обработка ответа
    if completion.choices:
        response_message = completion.choices[0].message
        if response_message.content:
            # Парсинг JSON-ответа в объект Entry
            parsed_response = Entry.parse_raw(response_message.content)
            print("Parsed response:", parsed_response)
        else:
            print("No content in the response.")
    else:
        print("No choices in the response.")
except Exception as e:
    print(f"Error: {e}")
```

```python
def split_text_in_memory(filepath, chunk_size_chars=1000, overlap_chars=200):
    """Читает весь файл в память и разбивает его на части с перекрытием."""
    try:
        with open(filepath, 'r', encoding='utf-8') as f:
            text = f.read()

        start = 0
        chunks = []
        while start < len(text):
            end = min(start + chunk_size_chars, len(text))
            
            # Находим последний перенос строки в пределах overlap
            if end + overlap_chars < len(text):
              overlap = text[end:end + overlap_chars]
              newline_pos = overlap.rfind('\n')

              if newline_pos != -1:
                end = end + newline_pos + 1
            
            chunk = text[start:end]
            chunks.append(chunk)
            start = end

        return chunks

    except FileNotFoundError:
        print(f"Файл {filepath} не найден.")
        return [] # Возвращаем пустой список в случае ошибки
    except Exception as e:
        print(f"Произошла ошибка: {e}")
        return []



# Пример использования:
filepath = "input_data/4pages.txt" # Замените на ваш путь
chunk_size = 1000
overlap = 200

chunks = split_text_in_memory(filepath, chunk_size, overlap)

for i, chunk in enumerate(chunks):
    print(f"Обработка части {i+1}:")
    # ... ваш код для обработки chunk ...
    print(chunk)
    print('-'*10)
    print("\n" * 4)
```
